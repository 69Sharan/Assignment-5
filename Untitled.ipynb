```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import influence_plot
```

```python
ran = pd.read_csv('ToyotaCorolla.csv', encoding='latin1')
ran
```

```python
ran.duplicated()
```

```python
ran.isnull().sum()
```

```python
ran2 = pd.concat([ran.iloc[:,2:4],ran.iloc[:,6:7],ran.iloc[:,8:9],ran.iloc[:,12:14],ran.iloc[:,15:18]],axis=1)
ran2
```

```python
ran3=ran2.rename({'Age_08_04':'Age','cc':'CC','Quarterly_Tax':'QT'},axis=1)
ran3
```

```python
features= ran3.drop('Price',axis=1)
features
```

```python
target = ran3[['Price']]
target
```

```python
x_train,x_test,y_train,y_test= train_test_split(features,target,train_size=0.7,random_state=10)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
lin_model= LinearRegression()
```

```python
lin_model
```

```python
lin_model.fit(x_train,y_train)
```

```python
y_pred= lin_model.predict(x_test)
```

```python
y_pred
```

```python
y_test
```

```python
err= y_test-y_pred
err
```

```python
plt.plot(y_pred, label='pred')
plt.plot(ran3, label='actual')
plt.legend()
plt.show()
```
