### Checking Collinearity & Residuals

```python
rsq_r=smf.ols("RD~ADM+MKTS",data=data1).fit().rsquared
vif_r=1/(1-rsq_r)

rsq_a=smf.ols("ADM~RD+MKTS",data=data1).fit().rsquared
vif_a=1/(1-rsq_a)

rsq_m=smf.ols("MKTS~RD+ADM",data=data1).fit().rsquared
vif_m=1/(1-rsq_m)

# Putting the values in Dataframe format
d1={'Variables':['RD','ADM','MKTS'],'Vif':[vif_r,vif_a,vif_m]}
Vif_df=pd.DataFrame(d1)
Vif_df
```

```python
# We found that None of the variable has VIF>20, Hence, there is No Collinearity. So we will consider all varaibles in Regression equation
```

```python
# Testing for Normality of Residuals i.e by (Q-Q Plot) using residual model (model.resid)

sm.qqplot(model.resid,line='q')
plt.title("Normal Q-Q plot of residuals")
plt.show()
```

```python
list(np.where(model.resid<-30000))
```

```python
def standard_values(vals) : 
    return (vals-vals.mean())/vals.std()  # z = (x - mu)/sigma
```

```python
plt.scatter(standard_values(model.fittedvalues),standard_values(model.resid))
plt.title('Residual Plot')
plt.xlabel('Standardized Fitted Values')
plt.ylabel('Standardized Residual Values')
plt.show() 
```

### Improving the Model

```python
# Discarding the data points which are influencers and reassigning the row number.

data2 = data1.drop(data1.index[[49]],axis=0).reset_index(drop=True)
data2
```

### Preparing another Model

```python
model2 = smf.ols("Profit~RD+ADM+MKTS",data=data2).fit()
```

```python
while model2.rsquared < 0.99:
    for c in [np.max(c)>1]:
        model2=smf.ols("Profit~RD+ADM+MKTS",data=data2).fit()
        (c,_)=model2.get_influence().cooks_distance
        c
        np.argmax(c) , np.max(c)
        data2=data2.drop(data2.index[[np.argmax(c)]],axis=0).reset_index(drop=True)
        data2
    else:
        final_model=smf.ols("Profit~RD+ADM+MKTS",data=data2).fit()
        final_model.rsquared , final_model.aic
        print("Thus model accuracy is improved to",final_model.rsquared)
```
